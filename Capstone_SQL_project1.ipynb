Importing libraries 

```python
import pandas as pd
df1=pd.read_csv("/content/Credit_card.csv")
df2=pd.read_csv("/content/Credit_card_label.csv")
data=df1.merge(df2,how='outer')           #merging two datassets
data.head()
```

```python
data.isnull().sum()
```

**observation:** As we can see data dataset contains null values.we need to deal with null values to perform sql queries.

The type_occupation column is not required for SQL queries.so I am going to drop type_occupation.

```python
data.drop(["Type_Occupation"],axis=1,inplace=True)
```

```python
data.head()
```

Here Iam imputing gender column with mode, Annual_income,Birthday_count with mean

```python
data["GENDER"].fillna(data["GENDER"].mode()[0],inplace=True)
data["GENDER"].isnull().sum()
data["Annual_income"].fillna(data["Annual_income"].mean(),inplace=True)
data["Birthday_count"].fillna(data["Birthday_count"].mean(),inplace=True)
data.isnull().sum()
```

Now this data set does not contain any null values.

By using pandas library i am converting the cleaned dataset in to csv file.

```python
data1=pd.DataFrame(data)
data1.to_csv("C:\\Users\\pc\\Desktop\\Dataset\\new.csv",index=False)
```
